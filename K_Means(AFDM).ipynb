```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from prince import FAMD
from sklearn.decomposition import PCA
```

```python
X_ini=pd.read_excel('DB_J(RRN,RTE500).xlsx')  
X=X_ini.iloc[680:1689,:] 
X=X.reset_index()
X=X.drop(['index'],axis=1)  
X=X.drop(['T2007','T2008','T2009','T2010','T2011','T2012','T2013','T2014','T2015'], axis=1) 
X=X.fillna('-999')
famd = FAMD(n_components =7, n_iter = 3, random_state = 101)
famd.fit(X)
X_KM=famd.transform(X)
X_KM=X_KM.reset_index()
X_KM=X_KM.drop('index', axis=1)
```

```python
############################ Application de KMeans ############################
kmeans = KMeans(n_clusters=6, precompute_distances='ward')
kmeans.fit(X_KM)
kmeans.predict(X_KM)
kmeans.labels_

# Index triés des groupes :  
idk = np.argsort(kmeans.labels_)
# Affichage des observations et leurs groupes :
Res_kmeans=pd.DataFrame(X_KM.index[idk],kmeans.labels_[idk])
```

```python
X_copy=X.copy()
X_copy=X_copy.drop(['longueur','concessionPrD','absD','cumulD','xD','yD','zD','prF','depPrF','absF','cumulF',
                    'latitudeF','xF','yF','zF','NumVoie','Classe-Etat_lib','Dept','Hierrarchie'], axis=1)
L=pd.DataFrame(columns=X_copy.columns)
for i in X_KM.index[idk]:
    j=i
    L=pd.concat([L,X_copy.iloc[i:j+1,:]], ignore_index=True)   
L['Clusters']=kmeans.labels_[idk]
L['index[idk]']=X_ini.index[idk]
KM_Clusters=L
```

```python
################ Profil moyen des clusters obtenues par K_means ###############
Clusters_counts=KM_Clusters['Clusters'].value_counts()
Data_analy=KM_Clusters.groupby(['Clusters']).mean()
Data_sum=KM_Clusters.groupby(['Clusters']).sum()
M=(KM_Clusters['Clusters'].value_counts()/KM_Clusters['Clusters'].count())*100


# Bitumen :  
for i in range(0,max(KM_Clusters['Clusters'])+1):
    D=KM_Clusters[KM_Clusters['Clusters']==i]
    b=(D['Bitumen'].value_counts()/D['Bitumen'].count())*100
    print(i,end = "\n")
    print(b)
    
# Nature de la chaussée :
for i in range(0,max(KM_Clusters['Clusters'])+1):
    D=KM_Clusters[KM_Clusters['Clusters']==i]
    p=(D['profil'].value_counts()/D['profil'].count())*100
    print(i,end = "\n")
    print(p)
    
# Etat de la chaussée :
for i in range(0,max(KM_Clusters['Clusters'])+1):
    D=KM_Clusters[KM_Clusters['Clusters']==i]
    c=(D['Classe-Etat'].value_counts()/D['Classe-Etat'].count())*100
    print(i,end = "\n")
    print(c)

# Vocation :
for i in range(0,max(KM_Clusters['Clusters'])+1):
    D=KM_Clusters[KM_Clusters['Clusters']==i]
    c=(D['vocation'].value_counts()/D['vocation'].count())*100
    print(i,end = "\n")
    print(c) 
    
# Type Carrefour :
for i in range(0,max(KM_Clusters['Clusters'])+1):
    D=KM_Clusters[KM_Clusters['Clusters']==i]
    c=(D['typecarrefour_lib'].value_counts()/D['typecarrefour_lib'].count())*100
    print(i,end = "\n")
    print(c)        
```

```python
# ACP :
pca = PCA(n_components=2)
pcs = pca.fit_transform(X_KM)

pc1_values = pcs[:,0]
pc2_values = pcs[:,1]
colormap=np.array(["red", "green", "blue", "yellow", "orange", "aqua", "magenta","brown","pink"])
sns.scatterplot(x=pc1_values, y=pc2_values, c=colormap[kmeans.labels_], s=50)
plt.xlabel("Dim 1 (20,63)")
plt.ylabel("Dim 2 (19,24)")
plt.grid(True)
plt.show()

    
#Visualisation : 
plt.figure(figsize=(80,10))
plt.scatter(X.longitudeD, X.latitudeD, c=colormap[kmeans.labels_], s=1000)
plt.show()
```
