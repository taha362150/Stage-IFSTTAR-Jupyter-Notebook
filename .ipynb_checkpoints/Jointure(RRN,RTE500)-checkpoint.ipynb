```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
RRN=pd.read_excel('RRN_Final.xlsx')
RTE500=pd.read_excel('Df_RTE500.xlsx')
```

```python
############### Jointure des deux bases de données RRN, RTE500 ###############
Df_J=pd.merge(left=RRN, right=RTE500, on='longitudeD', how='left')
Df_J=Df_J.sort_values(by='longitudeD',ascending=True)
plt.figure(figsize=(30,20))
sns.heatmap(Df_J.isna(),cbar=False)
```
