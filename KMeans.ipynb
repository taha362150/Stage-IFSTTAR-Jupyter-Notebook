```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.decomposition import PCA
import math
```

```python
X_ini=pd.read_excel('DB_J(RRN,RTE500).xlsx')
X_ini=X_ini.iloc[680:1689,:] 
X_ini=X_ini.reset_index()
X_ini=X_ini.drop(['index'],axis=1)
   
X=X_ini.drop(['T2007','T2008','T2009','T2010','T2011','T2012','T2013','T2014','T2015'], axis=1) 
#print(X)
print(X.shape)   
```

```python
########################### Encodage ##########################################
dict_Description={'2(Inf7)':0, '2(Sup7)':1, '2x1':2, '2x2':3, '4':4, '3(Inf10)':5, '3(Sup10)':6}
X['Description']=X['Description'].map(dict_Description)

for col in X.select_dtypes('object').columns:
        X.loc[:,col] = LabelEncoder().fit_transform(X[col])        
#print(X)
```

```python
########################### Standarisation ####################################
X=X.fillna('-999')
#X_KM=pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)
X=pd.DataFrame(StandardScaler().fit_transform(X), columns=X.columns)
#print(X)
```

```python
# ACP :
pca = PCA(n_components=None)
X_pca=pca.fit_transform(X)
print(pca.explained_variance_ratio_.cumsum())
X_pca=pd.DataFrame(X_pca)
```

```python
############ Détermination du choix du nombre de composante principale #########

# 1- Critère de Kalis-Saporta-Spinakis :
KSS=1-math.sqrt((1-X_pca.shape[1])/(1-X_pca.shape[0]))  #Coefficient de Kalis-Saporta-Spinakis
n_CP=0
for vap in pca.explained_variance_:  # Les valeurs propres associées à chaque composante principale
    if vap>KSS:
        n_CP=n_CP+1
print(n_CP)   # Nombre de composante principale à conservé (=18) 


# 2- Critère du Coude :
VAR_total=sum(pca.explained_variance_)
Vap_vartotal=pca.explained_variance_/VAR_total

plt.plot(range(1,len(Vap_vartotal)+1),Vap_vartotal)
plt.title("La méthode du Coude (pour ACP)")
plt.xlabel("nombre de composante principale")
plt.ylabel("Valeurs propres relatives")
plt.show()   # Le premier point d'inflexion est 18 ==> Conservation de 18 composantes principales
```

```python
X_KM=X_pca.copy()

#Détermination de la valeur optimale de k :
tab=[]
for i in range(2,10):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(X_KM)
    #kmeans.fit(X_pca)
    tab.append(kmeans.inertia_)
plt.plot(range(2,10),tab)
plt.title("La méthode Eblow")
plt.xlabel("nombre de cluster")
plt.ylabel("Inertie intra-classe")
plt.show()
```

```python
#Application de KMeans :
kmeans = KMeans(n_clusters=7)#, precompute_distances='ward')
print(kmeans.fit(X_KM))
print(kmeans.predict(X_KM))
print(kmeans.labels_)
print(kmeans.inertia_)
```

```python
# Index triés des groupes :  
idk = np.argsort(kmeans.labels_)
print(idk)

# Affichage des observations et leurs groupes :
Res_kmeans=pd.DataFrame(X_ini.index[idk],kmeans.labels_[idk])
#print(Res_kmeans)
```

```python
# ACP :
    
print(pca.explained_variance_ratio_.cumsum())

pca = PCA(n_components=n_CP)
pcs = pca.fit_transform(X)

pc1_values = pcs[:,0]
pc2_values = pcs[:,1]
colormap=np.array(["red", "green", "blue", "yellow", "black", "orange", "aqua", "magenta"])

######## Présentation des clusters dans les 2 axes factoriels de l'ACP #########
sns.scatterplot(x=pc1_values, y=pc2_values, c=colormap[kmeans.labels_], s=20) 


# Visualisation : ( La segmentation de la route )
colormap=np.array(["red", "green", "blue", "yellow", "black", "orange", "aqua", "magenta"])
plt.figure(figsize=(80,10))
plt.scatter(X_ini.longitudeD, X_ini.latitudeD, c=colormap[kmeans.labels_], s=500)
plt.show()
```

```python
X_copy=X_ini.copy()
X_copy=X_copy.drop(['longueur','concessionPrD','absD','cumulD','xD','yD','zD','prF','depPrF',
                    'absF','cumulF','latitudeF','xF','yF','zF',
                    'NumVoie','Classe-Etat_lib','Dept','Hierrarchie'], axis=1)
L=pd.DataFrame(columns=X_copy.columns)
for i in X_ini.index[idk]: 
    j=i
    L=pd.concat([L,X_copy.iloc[i:j+1,:]], ignore_index=True)
#print(L)   
L['Clusters']=kmeans.labels_[idk]
L['index[idk]']=X_ini.index[idk]
KM_Clusters=L
```
